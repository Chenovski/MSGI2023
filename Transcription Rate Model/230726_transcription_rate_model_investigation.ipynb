```python
from numpy import exp,arange
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
from IPython import display
```

```python
def tau_func(Ec,Er,Ei,Cc,Cr,RT,tau_max):
    a = Cr*exp(-Er/RT)+Cc*Cr*exp(-(Ec+Er+Ei)/RT)
    b = 1+Cc*exp(-Ec/RT)+a
    return tau_max*a/b
```

```python
Ei = -3.26
Cc = 0.063
Cr = 0.5 # tbd
RT = 1.98e-3 * 310
tau_max = 1 # tbd

xscale = 5
yscale = 5
dxy = 0.1

ec = arange(-xscale,xscale,dxy)
er = arange(yscale,-yscale,-dxy)
Ec,Er = meshgrid(ec, er) # grid of point

Tau = tau_func(Ec,Er,Ei,Cc,Cr,RT,tau_max) # evaluation of the function on the grid

im = imshow(Tau,cmap=cm.turbo,extent=[-xscale,xscale,-yscale,yscale]) # drawing the function
# adding the Contour lines with labels
cset = contour(ec,er,Tau,arange(-2,1,0.1),linewidths=1,cmap=cm.Paired)
clabel(cset,inline=True,fmt='%1.1f',fontsize=10)
colorbar(im) # adding the colobar on the right
title('transcription rate $\\tau$')
plt.xlabel('$\\varepsilon_r$')
plt.ylabel('$\\varepsilon_c$')
show()
```

x axis: RNAP energy $\varepsilon_r$, y axis: CRP energy $\varepsilon_c$.

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(Ec, Er, Tau, rstride=1, cstride=1, 
                      cmap=cm.turbo,linewidth=0, antialiased=False)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.view_init(azim=30, elev=30)
title('transcription rate $\\tau$')
plt.show()
```

```python
for Cr in arange(.01,1,.01):
    display.clear_output(wait=True);
    Tau = tau_func(Ec,Er,Ei,Cc,Cr,RT,tau_max) # evaluation of the function on the grid

    im = imshow(Tau,cmap=cm.turbo,extent=[-xscale,xscale,-yscale,yscale]) # drawing the function
    # adding the Contour lines with labels
    cset = contour(ec,er,Tau,arange(-2,1,0.1),linewidths=1,cmap=cm.Paired)
    clabel(cset,inline=True,fmt='%1.1f',fontsize=10)
    colorbar(im) # adding the colobar on the right
    title('transcription rate $\\tau$, when $C_r$ = %.2f' % Cr)
    plt.xlabel('$\\varepsilon_r$')
    plt.ylabel('$\\varepsilon_c$')
    show()
    plt.pause(.01)


Cr = 0.5;# set back
```
