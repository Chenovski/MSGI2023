```python
import numpy as np
import mptools as mp
import matplotlib.pyplot as plt
```

```python
ThetaRaw = mp.combTheta('data/theta230802bootstrap',4,10)
```

```python
ThetaRaw.shape
```

```python
Theta, mean, std, _, _ = mp.empiricalRule(ThetaRaw)
```

```python
fig, ax = plt.subplots(figsize=(4,3))

vp = ax.violinplot([Theta[:,1],Theta[:,2]], [2, 4], widths=1.5,showextrema=False)
# styling:
for body in vp['bodies']:
    body.set_alpha(0.6)
ax.errorbar([2, 4], mean[1:3], std[1:3], fmt='o', linewidth=2, capsize=3, color = 'black')
ax.set_xticks([2,4])
ax.set_xticklabels(['$C_c$','$C_r$'])
ax.xaxis.grid(False)
ax.yaxis.grid(linestyle='--')
ax.set(xlim=(0, 6), ylim=(.02, .06))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4,3))

vp = ax.violinplot(Theta[:,0], [3], widths=1.5,showextrema=False)
# styling:
for body in vp['bodies']:
    body.set_alpha(0.6)
ax.errorbar([3], mean[0], std[0], fmt='o', linewidth=2, capsize=3, color = 'black')
ax.set_xticks([3])
ax.set_xticklabels(['$\\bar{\\tau}/\\tau_{max}$'])
ax.xaxis.grid(False)
ax.yaxis.grid(linestyle='--')
ax.set(xlim=(0, 6), ylim=(.4, .5))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4,3))

vp = ax.violinplot(Theta[:,3], [3], widths=1.5,showextrema=False)
# styling:
for body in vp['bodies']:
    body.set_alpha(0.6)
ax.errorbar([3], mean[3], std[3], fmt='o', linewidth=2, capsize=3, color = 'black')
ax.set_xticks([3])
ax.set_xticklabels(['$\\varepsilon_i$'])
ax.xaxis.grid(False)
ax.yaxis.grid(linestyle='--')
ax.set(xlim=(0, 6), ylim=(-5.5,-5))

plt.show()
```
